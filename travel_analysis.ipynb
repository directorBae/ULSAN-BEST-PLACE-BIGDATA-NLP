```python
selected = ['장생포고래박물관', '울산대공원', '장생포고래문화마을', '태화강동굴피아',  '울산과학관', '고래생태체험관',  '장생포모노레일', '울산항만선착장', '문수야구장', '정토사', '태화강전망대', '태화강국가정원', '대왕암공원', '일산해수욕장', '주전몽돌해변', '슬도', '방어진활어센터', '주전항', '정자해수욕장', '강동몽돌해변', '태화강억새군락지', '진하해수욕장', '간절곶', '나사리해수욕장', '태화강생태관', '간월재', '솔개해수욕장', '울산옹기박물관', '백양사', '함월루', '학성공원', '태화루', '해남사', '약사동제방유적전시관',]
```

```python
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

from transformers import AutoTokenizer, AutoModelForSequenceClassification, TextClassificationPipeline

# hugging face에서 토크나이저와 모델 불러오기
tokenizer = AutoTokenizer.from_pretrained("searle-j/kote_for_easygoing_people")
model = AutoModelForSequenceClassification.from_pretrained("searle-j/kote_for_easygoing_people")

# 감정 분석 모델 설정 
pipe = TextClassificationPipeline(
        model=model,
        tokenizer=tokenizer,
        device=0 if torch.cuda.is_available() else -1,
        return_all_scores=True,
        function_to_apply='sigmoid'
    )
```

```python
result = pipe(up["content"].iloc[0], max_length=512, truncation=True, padding="max_length")

positive_score = 0.0
negative_score = 0.0

for r in result[0]:
    if r["label"] in ["감동/감탄", "편안/쾌적", '신기함/관심', '즐거움/신남', '흐뭇함(귀여움/예쁨)', '행복', '기쁨', '안심/신뢰']:
        positive_score += r["score"]
    elif r["label"] in ["안타까움/실망",  '절망', '어이없음', '짜증', '역겨움/징그러움', '힘듦/지침', '부담/안_내킴', '재미없음', '불안/걱정', '불쾌함/싫음', '슬픔/우울', '화남/분노', '무서움/두려움', '우울함/허탈함', '피곤함/지침']:
        negative_score += r["score"]

pos_avg = positive_score / 8
neg_avg = negative_score / 15
print(pos_avg, neg_avg)
print(pos_avg - neg_avg)
```

```python
from tqdm import tqdm
```

```python
pos_score_list_all = []
neg_score_list_all = []

for i in tqdm((range(int(len(up) / 100) + 1))):
    result_all = pipe(up["content"].iloc[i*100:(i+1)*100].tolist(), truncation=True, max_length=512, padding="max_length")
    pos_score_list = []
    neg_score_list = []

    for j in range(len(result_all)):
        positive_score = 0.0
        negative_score = 0.0
        for r in result_all[j]:
            if r["label"] in ["감동/감탄", "편안/쾌적", '신기함/관심', '즐거움/신남', '흐뭇함(귀여움/예쁨)', '행복', '기쁨', '안심/신뢰']:
                positive_score += r["score"]
            elif r["label"] in ["안타까움/실망",  '절망', '어이없음', '짜증', '역겨움/징그러움', '힘듦/지침', '부담/안_내킴', '재미없음', '불안/걱정', '불쾌함/싫음', '슬픔/우울', '화남/분노', '무서움/두려움', '우울함/허탈함', '피곤함/지침']:
                negative_score += r["score"]
            
        pos_avg = positive_score / 8
        neg_avg = negative_score / 15
        pos_score_list.append(pos_avg)
        neg_score_list.append(neg_avg)
    
    pos_score_list_all += pos_score_list
    neg_score_list_all += neg_score_list
```

```python
len(pos_score_list_all), len(neg_score_list_all)
```

```python
up["pos_score"] = pos_score_list_all
up["neg_score"] = neg_score_list_all
```

```python
up["total_score"] = up["pos_score"] - up["neg_score"]
```

```python
up.to_json("ulsan_place_score.json", orient="records", force_ascii=False, lines=True)
up.to_csv("ulsan_place_score.csv", index=False, encoding="utf-8-sig")
```

```python
## unique한 url에 대해 score 기준으로 평균값 계산
up_avg = up.groupby("url").agg({"pos_score": "mean", "neg_score": "mean", "total_score": "mean"}).reset_index()
```

```python
up_avg = up_avg.sort_values("total_score", ascending=False)
```

```python
## 각 url의 record 수 확인
up_avg["count"] = up.groupby("url").size().values
```

```python
up_avg
```

```python
## total_score와 count를 정규분포로 변환하고, Z-score를 계산
up_avg["total_score_z"] = (up_avg["total_score"] - up_avg["total_score"].mean()) / up_avg["total_score"].std()
up_avg["count_z"] = (up_avg["count"] - up_avg["count"].mean()) / up_avg["count"].std()

# Z-Score 합산 후 해당 값 기준으로 정렬
up_avg["z_score"] = up_avg["total_score_z"] + up_avg["count_z"]
up_avg = up_avg.sort_values("z_score", ascending=False)
```
