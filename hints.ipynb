# Hint 1 : Sentimental Analysis on dataframe 'up'

```python
# Make 'results' as a 1d array with maintaining the order
results = [item for sublist in results for item in sublist]
```

```python
assert len(results) == len(up)
```

```python
up["sentiment"] = results
```

```python
# up["url"] 기준으로 묶어서, 각 url에 대해 sentiment의 평균을 구함
# 이제 각 url에 대해 sentiment의 key별로 평균을 구합니다.

# 감정 라벨 목록을 추출합니다
all_labels = set()
for sentiment_list in up['sentiment']:
    for sentiment in sentiment_list:
        all_labels.add(sentiment['label'])

all_labels = list(all_labels)
print(f"감정 라벨 개수: {len(all_labels)}")

# 장소별로 감정 점수 평균을 저장할 빈 딕셔너리 생성
sentiment_avg_by_url = {}

# 각 장소별로 감정 점수 평균 계산
for url in up['url'].unique():
    # URL 필터링
    url_data = up[up['url'] == url]
    
    # 해당 URL의 sentiment 데이터 수집
    sentiment_scores = {label: [] for label in all_labels}
    
    for sentiment_list in url_data['sentiment']:
        for sentiment in sentiment_list:
            label = sentiment['label']
            score = sentiment['score']
            sentiment_scores[label].append(score)
    
    # 각 감정 라벨별로 평균 계산
    avg_scores = {}
    for label, scores in sentiment_scores.items():
        if scores:
            avg_scores[label] = sum(scores) / len(scores)
        else:
            avg_scores[label] = 0.0
    
    sentiment_avg_by_url[url] = avg_scores

# 결과를 데이터프레임으로 변환
sentiment_avg_df = pd.DataFrame.from_dict(sentiment_avg_by_url, orient='index')
sentiment_avg_df
```

```python
# 모든 장소의 평균 감정 점수 계산
overall_avg_sentiment = sentiment_avg_df.mean()
top_overall_sentiments = overall_avg_sentiment.sort_values(ascending=False).head(10)

print("울산 관광지 전체 상위 10개 감정:")
for label, score in top_overall_sentiments.items():
    print(f"  {label}: {score:.4f}")

# 시각화: 전체 관광지의 상위 10개 감정
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.rc('font', family='Malgun Gothic')
top_overall_sentiments.plot(kind='bar')
plt.title('울산 관광지 전체 상위 10개 감정')
plt.xlabel('감정')
plt.ylabel('평균 점수')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 각 장소별로 상위 5개 감정 차트 생성
import matplotlib.pyplot as plt
import numpy as np
import math

# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic')

# 그리드 설정 (32개 장소를 배치할 그리드)
n_places = len(sentiment_avg_df)
cols = 4  # 4개의 열
rows = math.ceil(n_places / cols)  # 필요한 행 수 계산

# 큰 figure 생성
plt.figure(figsize=(20, 5 * rows))

# 각 장소별로 상위 5개 감정 시각화
for idx, url in enumerate(sentiment_avg_df.index):
    # 상위 5개 감정 추출
    top5_sentiments = sentiment_avg_df.loc[url].sort_values(ascending=False).head(5)
    
    # 서브플롯 위치 계산
    plt.subplot(rows, cols, idx + 1)
    
    # 막대 그래프 생성
    bars = plt.bar(top5_sentiments.index, top5_sentiments.values, color='skyblue')
    
    # 막대 위에 값 표시
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                 f'{height:.2f}', ha='center', va='bottom', fontsize=9)
    
    # 축 레이블과 제목 설정
    plt.xticks(rotation=45, ha='right', fontsize=9)
    plt.ylim(0, 1.0)  # y축 범위 설정
    plt.title(f'{url}의 상위 5개 감정', fontsize=12)
    
    # y축 레이블 설정
    if idx % cols == 0:
        plt.ylabel('감정 점수', fontsize=10)

# 레이아웃 조정
plt.tight_layout()
plt.savefig('각_장소별_상위_5개_감정.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Hint 2: TOP 5 장소에 대한 워드클라우드

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import numpy as np
from konlpy.tag import Okt
import PIL.Image

# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic')

# 불용어 설정
stopwords = ['울산', '것', '등', '를', '이', '가', '은', '는', '에', '로', '에서', '했', '했어요', '있어요', '있습니다']

# 형태소 분석기
okt = Okt()

# 1. 전체 데이터로 워드클라우드 생성하기
def generate_wordcloud(text_data, additional_stopwords, title):
    """
    텍스트 데이터로부터 워드클라우드를 생성합니다
    """
    # 형태소 분석 및 명사 추출
    nouns = []
    for text in text_data:
        if isinstance(text, str):
            nouns.extend(okt.nouns(text))

    # 명사 리스트에서 불용어 제거
    # 불용어 리스트에 있는 단어는 제외
    current_stopwords = stopwords.copy()
    if additional_stopwords:
        current_stopwords.extend(additional_stopwords)

    # 불용어 제거 및 단어 카운트
    words = [word for word in nouns if word not in current_stopwords and len(word) > 1]
    word_counts = Counter(words)

    cloud = WordCloud(
        font_path='malgun',
        background_color='white',
        max_words=200,
        width=800, 
        height=600,
        colormap='viridis'
    ).generate_from_frequencies(word_counts)
    
    # 시각화
    plt.figure(figsize=(12, 8))
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis('off')
    plt.tight_layout()
    plt.savefig(f'wordcloud_{title}.png', dpi=300, bbox_inches='tight')
    
    plt.show()
    
    return word_counts

top_places = ["간절곶", "태화강억새군락지", "장생포고래문화마을", "일산해수욕장", "함월루"]
stopwords_add = {
    "간절곶": ["간절곶", "울산"],
    "태화강억새군락지": ["억새", "태화강", "군락", "군락지"],
    "장생포고래문화마을": ["장생포", "고래", "문화", "마을"],
    "일산해수욕장": ["일산", "해수욕장"],
    "함월루": ["함월루", "누각"]
}

for place in top_places:
    place_reviews = up[up['url'] == place]['content']
    generate_wordcloud(
        place_reviews,
        stopwords_add.get(place, []),
        title=place
    )
```
